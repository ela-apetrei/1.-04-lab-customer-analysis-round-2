```python
import pandas as pd
```

```python
mkt_analysis = pd.read_csv('marketing_customer_analysis.csv')
mkt_analysis
```

## Shaping the data

```python
mkt_analysis.shape
```

## Standardizing the columns

```python
print(mkt_analysis.columns)
```

```python
#lower cases
mkt_analysis.columns = mkt_analysis.columns.str.lower()
mkt_analysis
```

```python
# standardized column names will be meaninfull strings, lower cased and with '_' instead of 'spaces'

mkt_analysis.columns = [mkt_analysis.columns[i].lower().replace(' ', '_') for i in range(len(mkt_analysis.columns)) ]
 
mkt_analysis
```

```python
# renaming columns
mkt_analysis = mkt_analysis.rename(columns={'unnamed:_0':'id'})
mkt_analysis
```

## Datetime format
Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. _Hint_: If data from March does not exist, consider only January and February.

```python
mkt_analysis_clean_ver1 = pd.read_csv('mkt_analysis_clean_ver1.csv')
```

```python
mkt_analysis_clean_ver1['effective_to_date']
```

```python
mkt_analysis_clean_ver1['effective_to_date'] = pd.to_datetime(mkt_analysis_clean_ver1['effective_to_date'], errors='coerce')
mkt_analysis_clean_ver1['effective_to_date']
```

```python
mkt_analysis_clean_ver1["month"] = mkt_analysis_clean_ver1['effective_to_date'].map(lambda x: x.month)
mkt_analysis_clean_ver1.month
```
